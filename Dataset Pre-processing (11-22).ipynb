```python
def read_dataset(filepath):    
    df = pd.read_csv(filepath, header=None)
    df_n_row, df_n_column = df.shape
    df_columns = ["X{}".format(i) for i in range(0, df_n_column-1)] + ["y"]
    df.columns = df_columns
    df = df[df_columns]
    
    return df
```

```python
def save_df(df, filename, path):
    if not os.path.exists(path):
        os.mkdir(path)

    fullname = os.path.join(path, filename)
    df.to_csv(fullname, header=None, index=None, sep=',')
```

```python
classes = list(set([y[:y.rindex('-')] for y in os.listdir('../Bases/Auslan/tctodd1/')]))
subdirs = ['tctodd{}'.format(i) for i in range(1,10)]
samples = [i for i in range(1, 4)]

dfs = {}
for subdir in subdirs:
    for clazz in classes:
        if clazz not in dfs:
            dfs[clazz] = []
            
        for sample in samples:
            dfs[clazz].append(pd.read_csv('../Bases/Auslan/{}/{}-{}.tsd'.format(subdir, clazz, sample), sep='\t', header=None))

df_mean = {}
df_full = {}
for clazz in classes:
    df_mean[clazz] = pd.concat([df.mean() for df in dfs[clazz]], axis=1).T
    df_mean[clazz]['y'] = clazz
    
    df_full[clazz] = pd.concat(dfs[clazz])
    df_full[clazz]['y'] = clazz

def format_df(auslan):
    df = pd.concat(auslan.values())
    df_n_row, df_n_column = df.shape
    df_columns = ["X{}".format(i) for i in range(0, df_n_column-1)] + ["y"]
    df.columns = df_columns
    df = df[df_columns]
    
    return df
```

```python
auslan = format_df(df_mean)
display(auslan.head())
display(auslan.shape)
display(auslan.isnull().values.any())

# categorical to numerical
auslan['y'] = auslan['y'].astype('category')
cat_columns = auslan.select_dtypes(['category']).columns
auslan[cat_columns] = auslan[cat_columns].apply(lambda x: x.cat.codes)

auslan['y'] = [clazz+1 for clazz in list(auslan['y'])]
display(auslan['y'].unique())

display(auslan.head())

save_df(auslan, 'auslan.csv', './ProcessedBases/Auslan/')
```
