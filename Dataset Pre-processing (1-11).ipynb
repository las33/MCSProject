# Car

```python
current_base = 'car'

data = loadarff(basedir+current_base+'.dat')
df = pd.DataFrame(data[0])
df = df.applymap(lambda x: x.decode('utf-8')) 

cleanup_nums = {"Buying":   {"vhigh": 4, "high": 3, "med": 2, "low": 1 },
                "Maint":    {"vhigh": 4, "high": 3, "med": 2, "low": 1 },
                "Doors":    {"5more": 5},
                "Persons":  {"more": 6},
                "Lug_boot": {"small": 1, "med": 2, "big": 3 },
                "Safety":   {"low": 1, "med": 2, "high": 3 },
                "y":    {"unacc": 1, "acc": 2, "vgood": 4, "good": 3 }
               }

print("valores faltantes:", df.isnull().values.any())

df.rename(columns={'Acceptability':'y'}, inplace=True)
obj_df = df.replace(cleanup_nums)
df
```

```python
obj_df.to_csv(savedir+current_base+'.csv', index = False)
obj_df
```

# Lymphography

```python
current_base = 'lymphography'

data = loadarff(basedir+current_base+'.dat')
df = pd.DataFrame(data[0])

clear_names = {'Lymphatics', 'Block_of_affere', 'Bl_of_lymph_c', 'Bl_of_lymph_s',
              'By_pass', 'Extravasates', 'Regeneration_of' ,'Early_uptake_in',
              'Changes_in_lym',  'Defect_in_node', 'Changes_in_node', 'Changes_in_stru',
              'Special_forms', 'Dislocation_of', 'Exclusion_of_no', 'Class'}

for name in clear_names: 
    df[name] = df[name].apply(lambda x: x.decode('utf-8')) 

print("valores faltantes:", df.isnull().values.any())

df.rename(columns={'Class':'y'}, inplace=True)

df
```

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
        
df.to_csv(savedir+current_base+'.csv', index = False)
df
```

# Autos

```python
current_base = 'autos'

data = loadarff(basedir+current_base+'.arff')
df = pd.DataFrame(data[0])
print("valores faltantes:", df.isnull().values.any())
df = df.dropna()    
    
print("valores faltantes:", df.isnull().values.any())
#df.rename(columns={'Class':'y'}, inplace=True)

df
```

```python

clear_names = {'make','symboling', 'fuel-type', 'aspiration', 'num-of-doors',
              'body-style', 'drive-wheels', 'engine-location' ,'engine-type',
              'num-of-cylinders',  'fuel-system'}

for name in clear_names: 
    df[name] = df[name].apply(lambda x: x.decode('utf-8')) 
    
cleanup_nums = {"num-of-doors":   {"four": 4, "two": 2 },
                "num-of-cylinders":    {"eight": 8, "five": 5, 
                                        "four": 4, "six":6 , "three": 3, "twelve": 12, "two":2}} 
    

cat = {'make','symboling', 'fuel-type', 'aspiration',
              'body-style', 'drive-wheels', 'engine-location' ,'engine-type',  'fuel-system'}
    
df = df.replace(cleanup_nums)


for col_name in cat:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes

df.rename(columns={'symboling':'y'}, inplace=True)
df.to_csv(savedir+current_base+'.csv', index = False)
df
```
